# Amazon Product and its Ratings

```python
#importing required libraries
import numpy as np
import pandas as pd
pd.pandas.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('Amazon_Products.csv', low_memory=False)
df.head()
```

```python
#dropping columns with nan values
df.drop(df.columns[14:],axis=1,inplace=True)
df.drop(df.columns[[9]],axis=1,inplace=True)
df.head()
```

```python
df['average_review_rating'].isnull().sum()
```

```python
#to separate the rating from the text in order to fill nan values
new=df['average_review_rating'].str.split(" ", n = 3, expand = True)
df['average_review_rating']=new[0]
```

```python
df.head()
```

```python
df['average_review_rating'].unique()
```

```python
#converting dtype from object to float
df['average_review_rating']=pd.to_numeric(df['average_review_rating'],errors='coerce')
```

```python
#since the data is MCAR(missing completely at random) we can apply mean imputation
mean=df['average_review_rating'].mean()
df['average_review_rating'].fillna(mean)
df.head()
```

```python
#removing rows having nan values by simply dropping it
df.dropna(axis=0,inplace=True)
```
